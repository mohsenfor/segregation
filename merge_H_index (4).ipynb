```python
# global H index for multi family/single family segregaton with formula from: Barros and Feitusa (2018)
# based on theils entropy with this original formula: H = Σ (pi * ln(pi / pavg))
# hierarchal structure of data: neigh_ID --> district_ID --> OSTAN
# we have three types of segregation: multifam/singlefam houses - rental/owned houses and also native/immigrant dwellers

# Step 1: Calculate total population (N) in the study area
total_population = data[['multi_family', 'single_family']].sum().sum()

# Step 2: Calculate proportion of each group (τm) in the study area
proportions_study_area = data[['multi_family', 'single_family']].sum() / total_population

# Step 3: Calculate total entropy of the study area (E)
E = np.sum(proportions_study_area * np.log2(1 / proportions_study_area))

# Step 4: Calculate entropy of each locality (Ej)
# Group data by district and neighborhood
grouped_by_district_neighborhood = data.groupby(['OSTAN', 'district_ID', 'neigh_ID'])

# Calculate Ej for each locality
Ej_values = []
for (ostan, district_id, neigh_id), group_data in grouped_by_district_neighborhood:
    # Calculate total population in the locality (Nj)
    total_population_locality = group_data[['multi_family', 'single_family']].sum().sum()
    
    # Calculate proportion of each group (τjm) in the locality
    proportions_locality = group_data[['multi_family', 'single_family']].sum() / total_population_locality
    
    # Calculate entropy of the locality (Ej)
    Ej = np.sum(proportions_locality * np.log2(1 / proportions_locality))
    Ej_values.append(Ej)

# Step 5: Calculate global H index
Nj_values = data.groupby(['OSTAN', 'district_ID', 'neigh_ID'])[['multi_family', 'single_family']].sum().sum(axis=1)
global_H_index = np.sum(Nj_values * (E - Ej_values)) / total_population

print("Global H Index:", global_H_index)
```

```python
# to calculate entropy at neighborhood scale with shannons entropy
# with this formula: H(p) = - Σ (pi * log2(pi))


# Define function to calculate entropy
def calculate_entropy(proportions):
    # Clip proportions to avoid log of 0 or 1
    epsilon = 1e-10
    proportions = proportions.clip(epsilon, 1 - epsilon)

    # Calculate entropy using proportions
    entropy = -np.sum(proportions * np.log2(proportions))

    return entropy

# Group data by OSTAN, district_ID, and neigh_ID
grouped_by_location = data.groupby(['OSTAN', 'district_ID', 'neigh_ID'])

# Initialize dictionary to store entropies
entropies_dict = {}

# Iterate through each location group
for (ostan, district_id, neigh_id), group_data in grouped_by_location:
    # Calculate proportions for each category
    proportions_owned_rent = group_data[['multi_family', 'single_family']].sum() / group_data[['multi_family', 'single_family']].sum().sum()
    proportions_native_migration = group_data[['native_2', 'migration']].sum() / group_data[['native_2', 'migration']].sum().sum()
    proportions_householders_renters = group_data[['householders', 'rent']].sum() / group_data[['householders', 'rent']].sum().sum()

    # Calculate entropies for each category
    entropy_owned_rent = calculate_entropy(proportions_owned_rent)
    entropy_native_migration = calculate_entropy(proportions_native_migration)
    entropy_householders_renters = calculate_entropy(proportions_householders_renters)

    # Store entropies along with OSTAN and district_ID in the dictionary
    entropies_dict[(ostan, district_id, neigh_id)] = {
        'OSTAN': ostan,
        'district_ID': district_id,
        'neigh_ID': neigh_id,
        'entropy_multi_single': entropy_owned_rent,
        'entropy_native_migration': entropy_native_migration,
        'entropy_householders_renters': entropy_householders_renters
    }

# Convert dictionary to DataFrame
entropies_df0 = pd.DataFrame.from_dict(entropies_dict, orient='index')

print (entropies_df0)
```
